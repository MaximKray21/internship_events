# Решение на питоне

```python
class Solution:
    
    def __init__(self):
        self.ans = 0
        
    def readdata(self, filename):
        with open(filename, 'r') as f:
            self.n = int(f.readline())
            self.nums = list(map(int, f.readline().split()))
        self.additional_arr = [[0] * self.n for _ in range(self.n)]
        
    def first_loop(self):
        for i in range(self.n):
            k = i
            count = 0
            while (k < self.n):
                if self.nums[i] < self.nums[k]:
                    count += 1
                self.additional_arr[i][k] = count
                k += 1
            k = i-1
            count = 0
            while k >= 0:
                if self.nums[k] < self.nums[i]:
                    count += 1
                self.additional_arr[i][k] = count
                k -= 1
    
    def sec_loop(self):
        self.denominator = 0
        p_z = self.n
        while p_z > 0:
            self.denominator += p_z - 1
            p_z -= 1
            
    def third_loop(self):
        self.start_counter = 0
        for i in range(self.n):
            for j in range(i+1, self.n):
                if self.nums[i] > self.nums[j]:
                    self.start_counter += 1
                    
    def four_loop(self):
        for i in range(self.n):
            for j in range(i+1, self.n):
                if self.nums[i] > self.nums[j]:
                    self.ans -= 1
                    self.ans += self.start_counter + (self.additional_arr[i][j] - (j - i - 1 - self.additional_arr[i][j])) + \
                        (self.additional_arr[j][i] - (j - i - 1 - self.additional_arr[j][i]))
                else:
                    self.ans += 1
                    self.ans += self.start_counter + (self.additional_arr[i][j] - 1 - (j - i - 1 - (self.additional_arr[i][j] - 1))) + (
                        self.additional_arr[j][i] - 1 - (j - i - 1 - (self.additional_arr[j][i] - 1)))
    
    def drob(self):
        answer_counter = 2
        while (answer_counter != self.denominator+1):
            if (self.ans % answer_counter == 0 and self.denominator % answer_counter == 0):
                self.ans = self.ans//answer_counter
                self.denominator = self.denominator//answer_counter
                answer_counter = 1
            answer_counter += 1
        return str(self.ans) + '/' + str(self.denominator)

solver = Solution()
filename = 'input.txt'
solver.readdata(filename)
solver.first_loop()
solver.sec_loop()
solver.third_loop()
solver.four_loop()
print(solver.drob())
```
