```python
import numpy as np
from sklearn.metrics import mean_squared_error 
from scipy.optimize import minimize

with open('input.txt', 'r') as f:
    data = f.read().splitlines()
    
number_of_views = data[0]
x_list, y_arr_true = [], []

for stroke in data[1:]:
    x, y = map(float, stroke.split())
    x_list.append(x), y_arr_true.append(y)

x_list = np.array(x_list)
y_arr_true = np.array(y_arr_true)
def ret_x(x, a_, b_, c_, d_):
    return (a_ * np.tan(x)) + ((b_ * np.sin(x) + c_ * np.cos(x)) ** 2) + (d_ * (x ** 0.5))

def minimizing_function(w):
    a, b, c, d = w
    global x_list, y_list
    arr_y = ret_x(x_list, a, b, c, d)
    return mean_squared_error(y_true = y_arr_true, y_pred = arr_y)

result = minimize(minimizing_function, np.array([1, 1, 1, 1]), method='Powell')
print(*result.x)
```
